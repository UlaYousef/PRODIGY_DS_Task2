```python
def impute_Benefits(cols):
    Benefits = cols[0]
    TotalPayBenefits = cols[1]
    TotalPay = cols[2]
    if pd.isnull(Benefits):

        if TotalPayBenefits == TotalPay :
            return 0.0
    else:
        return Benefits
```

```python
df['Benefits'] = df[['Benefits','TotalPayBenefits','TotalPay']].apply(impute_Benefits,axis=1)
```

```python
df.loc[df['Benefits'].isnull()==True]
```

```python
df.loc[df['BasePay'].isnull()==True]
```

looking at previous dataframe we notice the following: TotalBay is the sum of BasePay,OvertimePay,OtherPay

TotalPayBenefits is the sum of TotalBay and Benefits

when BasePay has NaN value the TotalPay is equal to OtherPay, the OvertimePay is equal to zero

so BasePay is equal to zero


```python
def impute_BasePay(cols):
    BasePay = cols[0]
    TotalPay = cols[1]
    OtherPay = cols[2]
    OvertimePay= cols[3]
    if pd.isnull(BasePay):
        if OtherPay == TotalPay and OvertimePay==0.0:
            return 0.0
    else:
        return BasePay
```

```python
df['BasePay'] = df[['BasePay','TotalPay','OtherPay','OvertimePay']].apply(impute_BasePay,axis=1)
```

```python
df.loc[df['BasePay'].isnull()==True]
```

there are missing values still in BasePay column, but we notice from above that TotalPay is just the sum of OvertimePay and OtherPay so BasePay is also equal to zero

```python
def impute_BasePay2(cols):
    BasePay = cols[0]
    TotalPay = cols[1]
    OtherPay = cols[2]
    OvertimePay= cols[3]
    if pd.isnull(BasePay):
        if TotalPay == OtherPay + OvertimePay:
            return 0.0
    else:
        return BasePay
```

```python
df['BasePay'] = df[['BasePay','TotalPay','OtherPay','OvertimePay']].apply(impute_BasePay2,axis=1)
```
